```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_am = pd.read_csv("strains_am.csv",index_col=0)
df_ao = pd.read_csv("strains_ao.csv",index_col=0)
df_an = pd.read_csv("strains_an.csv",index_col=0)
df_ae = pd.read_csv("strains_ae.csv",index_col=0)

df_am_chisq = pd.read_csv("/u/home/r/rwolff/strain_stability_revisions/strainstability/analysis/chisq/Poyet/am_strain_chisq_test.txt",index_col=0,header=None)
df_ao_chisq = pd.read_csv("/u/home/r/rwolff/strain_stability_revisions/strainstability/analysis/chisq/Poyet/ao_strain_chisq_test.txt",index_col=0,header=None)
df_an_chisq = pd.read_csv("/u/home/r/rwolff/strain_stability_revisions/strainstability/analysis/chisq/Poyet/an_strain_chisq_test.txt",index_col=0,header=None)
df_ae_chisq = pd.read_csv("/u/home/r/rwolff/strain_stability_revisions/strainstability/analysis/chisq/Poyet/ae_strain_chisq_test.txt",index_col=0,header=None)
```

```python
df_am_chisq < .05
```

```python
#df_am = df_am/df_am.sum()
#df_ao = df_ao/df_ao.sum()
#df_an = df_an/df_an.sum()
#df_ae = df_ae/df_ae.sum()

df_am = df_am.dropna(axis=1,how="all")
df_an = df_an.dropna(axis=1,how="all")
df_ao = df_ao.dropna(axis=1,how="all")
df_ae = df_ae.dropna(axis=1,how="all")
```

### Supplemental figure broken down by host

```python
pval_thresh = 0.05

fig,axs = plt.subplots(2,2,figsize=(12,8))

axs = axs.ravel()

means = df_am.T.mean()
var = df_am.T.var()

for strain in df_am_chisq.index:
    if df_am_chisq.loc[strain].values[0] < pval_thresh:
        axs[0].scatter(means.loc[strain],var.loc[strain],color="red")
    else:
        axs[0].scatter(means.loc[strain],var.loc[strain],color="blue")

means = df_ao.T.mean()
var = df_ao.T.var()

for strain in df_ao_chisq.index:
    if df_ao_chisq.loc[strain].values[0] < pval_thresh:
        axs[1].scatter(means.loc[strain],var.loc[strain],color="red")
    else:
        axs[1].scatter(means.loc[strain],var.loc[strain],color="blue")

means = df_an.T.mean()
var = df_an.T.var()

for strain in df_an_chisq.index:
    if df_an_chisq.loc[strain].values[0] < pval_thresh:
        axs[2].scatter(means.loc[strain],var.loc[strain],color="red")
    else:
        axs[2].scatter(means.loc[strain],var.loc[strain],color="blue")

means = df_ae.T.mean()
var = df_ae.T.var()

for strain in df_ae_chisq.index:
    if df_ae_chisq.loc[strain].values[0] < pval_thresh:
        axs[3].scatter(means.loc[strain],var.loc[strain],color="red")
    else:
        axs[3].scatter(means.loc[strain],var.loc[strain],color="blue")

for i in range(4):
    axs[i].loglog()
    axs[i].set_ylim([1e-6,1])
    axs[i].set_xlim([1e-3,1])
```

### Alpha diversity

```python
fig,axs = plt.subplots(2,2,figsize=(12,8))

axs = axs.ravel()

axs[0].plot((-df_am*np.log(df_am)).sum().values);
axs[1].plot((-df_ao*np.log(df_ao)).sum().values);
axs[2].plot((-df_an*np.log(df_an)).sum().values);
axs[3].plot((-df_ae*np.log(df_ae)).sum().values);
```
